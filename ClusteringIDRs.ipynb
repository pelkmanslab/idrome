# Systematic analysis of the functionality of related IDRs in the human and viral proteomes 

The high prevalence of IDRs in important regulatory, signaling, and cell cycle machinery suggests that they serve functionally importmant roles in these critical cell processes. However, there has not been a systematic analysis of functional enrichment of IDRs with similar compositional biases.

```python
import pandas as pd, numpy as np
from mischelperfuncs import *
import seaborn as sns 
from collections import Counter
%matplotlib inline

human_proteome = pd.read_csv('gw_human_reviewed_beta.csv', index_col=0)[:-1]
viral_proteome = pd.read_csv('outputs/humanviruses_completeprots_beta_withpolyprots.csv', index_col=0)

human_proteome.Sequence = human_proteome.Sequence.str.replace(' ', '')
viral_proteome.Sequence = viral_proteome.Sequence.str.replace(' ', '')

proteomes = [human_proteome, viral_proteome]
proteome_names = ['Human', 'Viral']
```

Note: Since many of the same analyses are carried out on both the viral proteomes. I have adopted the convention of writing `proteomes[idx]` where `idx` is the index of the proteome of interest (0 = human, 1 = human viral). For example:

## Collect disordered sequences

```python
from mischelperfuncs import *

disordered_seq_gaps = []
disordered_seqs = [[] for proteome in proteomes]
for idx, proteome in enumerate(proteomes):
    gaps = []
    for row in proteome['Disordered Threshold: 30']:
        if pd.notnull(row):
            lims = np.array([int(lim) for idr in [pair.split('_') for pair in row.split(';')] for lim in idr])
            gaps.append(np.diff(lims)[1::2]-1)
    disordered_seq_gaps.append([gap for gaplist in gaps for gap in gaplist])
    disordered_seqs[idx] = proteome.apply(retrieve_disordered_regions, axis=1)
```

## Computing background frequencies

There are two different types of background:

1. True background (all sequences) referred to as `background_freqs` containing these values for both human and human viral proteomes.
2. Disordered background (all disordered sequences) referred to as `disordered_freqs` containing these values for both proteomes

```python
disordered_freqs = [[] for proteome in proteomes]
background_freqs = [[] for proteome in proteomes]
for idx, proteome in enumerate(proteomes):
    print("\n%s proteome:"%proteome_names[idx])
    disordered_freqs[idx] = get_freqs(''.join(disordered_seqs[idx][disordered_seqs[idx].notnull()]))
    background_freqs[idx] = get_freqs(''.join(proteome['Sequence'].tolist()))
```

## Fold enrichments in disordered sequences versus background

```python
for idx, proteome in enumerate(proteomes):
    print("\n%s proteome:"%proteome_names[idx])
    print(get_fold_enrichments(disordered_freqs[idx], background_freqs[idx]))
```

## Construct list of IDRs for both proteomes

```python
aas = get_aas()

disordered_df = [[] for proteome in proteomes]
idrs = []

for prot_idx, proteome in enumerate(proteomes):
    print("\n%s proteome:"%proteome_names[prot_idx])

    s = pd.DataFrame({"Seqs":disordered_seqs[prot_idx], 'Length':proteome['Length']})
    s = s[s['Seqs'].notnull()]

    split_disordered_prots = {'Entry':[], 'Sequence':[], 'Length':[]}
    split_disordered_prots.update({aa:[] for aa in aas})
    for prot, row in s.iterrows():
        sep_idrs = row['Seqs'].split('_')

        for idx, seq in enumerate(sep_idrs):
            split_disordered_prots['Entry'].append("%s_d%s"%(prot, idx+1))
    #         print(len(split_disordered_prots['Entry']))
            split_disordered_prots['Sequence'].append(seq)
            freqs = get_freqs(seq, verbose=False)
    #         print("%s | %s --> %s --> %s"%(split_disordered_prots['Entry'][-1], seq, freqs, [freqs[aa] for aa in aas]))
            split_disordered_prots['Length'].append(len(seq))
            [split_disordered_prots[aa].append(freqs[aa]) for aa in aas]

    # split_disordered_prots
    idrs.append(pd.DataFrame(split_disordered_prots, columns=['Entry', 'Sequence', 'Length']+[aa for aa in aas]))
    print(len(idrs[prot_idx].loc[:,'A':'Y'].values))
    
    disordered_df[prot_idx] = s
```

# Cluster Analysis

Now that we have the clusters, how do we determine which clusters are meaningful? The nested nature of hierarchical clusters makes this difficult, but one option to start with clusters of a minimum size $k$ and calculate the enrichment or depletion of GO terms at this level and then go up level so that the size is $2k$ and then run these tests again. Now, this technique has some major drawbacks and is not very robust (i.e. error due to multiple testing), but it should suffice for now. This should be a common problem in biological clustering so I'm surprised by the lack of literature on this topic (aside from a couple papers on analyzing microarray expression profiles).

Add support for going up the tree and for querying the leaves efficiently at each level in the tree

```python
t_leaves = []

# Yes, recursion is horribly inefficient in python. This should be feasible
# for smaller trees. 
def build_parents(parent, t, t_leaves):
    """
    Adds parent information to the hierarchy tree. ``t`` is the threshold
    for the minimum number of elements needed to be called a leaf. Leaf
    nodes will be appended to ``t_leaves`` 
    """
    l,r = parent.get_left(), parent.get_right()
    if l is not None:
        l.parent = parent
        build_parents(l, t, t_leaves)
    if r is not None:
        r.parent = parent
        build_parents(r, t, t_leaves)
    # ugh
    if getattr(l, 'count', None) < t and getattr(r, 'count', None) < t and parent.get_count() >= t:
        t_leaves.append(parent)

def build_child_info(parent):
    """
    For each node adds information of the leaves of original data contained in the
    subtree with that node as the root.
    """
    l,r = parent.get_left(), parent.get_right()
    if l is None and r is None:
        parent.children = []
        return [parent], [parent.id]
    left_children, left_ids = build_child_info(l)
    right_children, right_ids = build_child_info(r)
    parent.children = left_children + right_children
    parent.child_ids = left_ids + right_ids
    return parent.children, parent.child_ids

root.parent = root
build_parents(root, 20, t_leaves)
build_child_info(root);
```

```python
import scipy.stats as stats
import itertools

background_counts = get_go_counts(idrs[0]['Entry'], human_proteome)
# print(background_counts.most_common(50))
leaf_max_dist = []
szs = []

output = {}

for leaf in t_leaves:
    leaf = leaf.parent.parent.parent.parent
    
    if leaf.id in output.keys():
        continue
    
    go_counts = get_go_counts(idrs[0].iloc[leaf.child_ids]['Entry'], human_proteome)
    
    sz = len(leaf.child_ids)
    
    if sz > 10000: continue
#     print(sz)
    
    results = []
    
    for term, count in go_counts.most_common():
        if count <= 10: continue
#         go_terms = human_proteome.loc[df.Entry, :]['Gene ontology (GO)']
        sig = stats.hypergeom.sf(count, len(idrs[0]), background_counts[term], sz)    
        if sig < 0.005:
            bkgc = background_counts[term]
            results.append({'go_term':term, 'sig':sig, 'count':count, 'bkgc':bkgc, 'per':int(count/float(bkgc)*1000)})
    results.sort(key=lambda x: x['sig'])
#     print(results[0])
    if len(results) > 0:
        results[0].update({'subtree_sz':sz})
#         if 'O-glycan' in results[0]['go_term'] and sz < 250:
#             print(idrs[0].iloc[leaf.child_ids]['Entry'])
        leaf_max_dist.append(pdist(idrs[0].iloc[leaf.child_ids].loc[:, 'A':'Y']).max())
        szs.append(sz)
        output.update({leaf.id:results[0]})
df = pd.DataFrame(output.values(), index=output.keys(), columns=['subtree_sz', 'go_term', 'sig', 'count', 'bkgc', 'per'])
```

## Variation in AA frequency along IDR length

```python
deviations = np.zeros(len(idrs[0]))

for idx, idr in enumerate(idrs[0].Sequence):

    full_freq = np.array
    
    full_freq = get_freqs(idr, verbose=False)
#     print(full_freq)

    full_freq = np.array([full_freq[aa] for aa in aas])
    output = np.zeros(len(idr)-30)
#     print(len(output))

    for i in range(15, len(idr)-15):
        freqs = get_freqs(idr[i-15:i+15], verbose=False)
        diff = np.array([freqs[aa] for aa in aas]) - full_freq
        output[i-15] = np.sqrt(np.sum(diff**2, 0))
#         if output[i-15] > 0.47:
#             print("\n%s %s"%(260+i, freqs))
    deviations[idx] = output.max() if len(output) > 0 else 0
    
idrs[0]['Deviation_From_Mean'] = deviations

print(len(output))
```
