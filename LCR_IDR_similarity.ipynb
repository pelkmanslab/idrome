Low Complexity Regions (LCRs) and Intrinsically Disordered Regions (IDRs) are often present in the same protein. This bring up some interesting questions:

1. Is there overlap between LCRs and IDRs?
2. What is the percent overlap?
3. Is the overlap significant?
4. Are there certain kinds of AAs whose LCRs overlap more?

```python
import pandas as pd, numpy as np
import seaborn as sns
from __future__ import print_function
%pylab inline
import json
config = json.load(open('config.json'))
human_proteome = pd.read_csv(config['rawdata']['human'], index_col=0)
```


## Is there overlap?

A rough answer for the first question is fairly easy to achieve by plotting the overlap of proteins containing LCRs (as predicted by CAST) with those containing IDRs (as predicted by IUPred) that are 30 AAs or longer. A venn diagram would be perfect (requires [matplotlib_venn](https://github.com/konstantint/matplotlib-venn) package)

```python
from matplotlib_venn import venn2

idr_key = 'IDR30'
lcr_containing_prots = human_proteome[human_proteome['LCRs'].notnull()]
idr_containing_prots = human_proteome[human_proteome[idr_key].notnull()]

# Get proteins that have both lcrs and idrs
lcr_and_idrs = lcr_containing_prots[lcr_containing_prots[idr_key].notnull()]

venn2([set(lcr_containing_prots.index), 
       set(idr_containing_prots.index)], set_labels=('LCRs', 'IDRs'));
plt.gcf().savefig(config['output']['plots']+'lcr_idr_venn.png', transparent=True)
```

## What is the percent overlap?

The previous venn diagram is nice, but it is missing a vital piece of information: whether LCRs and IDRs are present in the same regions of the proteins. Presence in the same region would suggest an overlap in functionality. 

The percent overlap can be computed as follows:

- Retrieve the positions that are considered instrinically disordered and belong to regions that have more than 30 AAs in a row. Pool these together.
- Retrieve the positions of residues that belong to low complexity regions. Pool these together as well.
- Compute the percentage of the low complexity positions that fall within the intrinsically disordered positions using set intersection. 

```python
from mischelperfuncs import *
import re
import ast

query = re.compile(r'([A-Z]).*?(\[.*?\]);')
percent_overlaps = []
prot_lens = []
prot_lcrs = []
prot_idrs = []

for idx, row in lcr_and_idrs.iterrows():
    lcrs = []
    # Extract LCR positions
    for lcr in re.findall(query, row['LCRs']):
        # pool them
        lcrs += ast.literal_eval(lcr[1]) 
    prot_lcrs.append(lcrs)
    # Extract IDR starts and ends
    idrs = [tuple(idr.split('_')) for idr 
            in row['IDR30'].split(';')]
    idr_pos = set()
    # Compute IDR positions and pool them
    [idr_pos.update(range(int(idr[0]), int(idr[1])+1)) for idr in idrs]
    prot_idrs.append(idr_pos)
    prot_lens.append(len(row['SEQ'].replace(' ', '')))
    # compute set intersection and divide by total number of lcr positions
    itrsct = len(set(lcrs).intersection(idr_pos))/float(len(lcrs))*100
    percent_overlaps.append(itrsct)

# Plotting the histogram and kde
with sns.axes_style('ticks'):
    f = plt.figure(figsize=(8,5))
    ax = f.add_subplot(111)
    sns.despine(ax=ax)
    sns.offset_spines(ax=ax)
    sns.distplot(np.array(percent_overlaps), 
                 axlabel='LCR-IDR percent overlap', ax=ax);
    f.subplots_adjust(bottom=0.15)
    f.savefig(config['output']['plots']+'lcr_idr_overlap.png', transparent=True)
```

As expected, most LCRs are found within IDRs. But there are quite a few LCRs that do not overlap with IDRs. This then leads into questions 3 and 4.

## Is the overlap significant?

For each protein we have numbers $n_{LCRs}$, $n_{IDRs}$, and $n_{overlap}$ of residues classified as either part of a low complexity region, or an intrinsically disordered region, or both, respectively. The significance of this overlap in comparison to the total length of the protein $N$ can be determined via the hypergeometric function: 

$$ P(X=k) = {{{n_{LCRs} \choose n_{overlap}} {{N-n_{LCRs}} \choose {n_{IDRs}-n_{overlap}}}}\over {N \choose n_{overlap}}} $$

This value will be computed per protein. However, we are interested with the significance across the human proteome. Since each of the previous hypergeometric tests are independent and have the same null hypothesis, we can use [Fisher's method](http://en.wikipedia.org/wiki/Fisher%27s_method) to combine the p-values:

$$ X^2_{2k} \sim -2 \sum_{i=1}^k \ln p_i $$

where $X^2$ is a combined test statistic, $p_i$ is the the p-value for the $i$th hypothesis test, and $k$ is the number of tests being combined. The test statistic $X^2$ has $2k$ degrees of freedom. 

```python
from scipy.stats import chi2, hypergeom
import warnings

# check that everything has the same length
assert all([len(prot_lcrs) == len(prot_idrs), 
            len(prot_idrs) == len(prot_lens)]), "Discrepency in lengths"

pvals = np.zeros(len(prot_lcrs))

for idx, tup in enumerate(zip(prot_lcrs, prot_idrs, prot_lens)):
    lcr_poses, idr_poses, prot_len = tup
    overlap = len(set(lcr_poses).intersection(idr_poses))
    pvals[idx] = hypergeom.sf(overlap-1, len(lcr_poses)+len(idr_poses), 
                              len(lcr_poses), len(idr_poses))
    
# Fisher's method
warnings.warn("This calculation is incorrect, p-value cannot equal 0.")
x2 = -2*np.sum(np.log(pvals))
deg = 2 * len(pvals)
chi2.sf(x2, deg)
# np.sum(pvals)
```

## Does the type of LCR have an effect?

It is very likely that LCRs comprised of different amino acids would differentially favor overlap with IDRs. LCRs of amino acids that are enriched in IDRs, e.g. proline, will most likely be more prone to overlap than those that are depleted in IDRs, e.g. tryptophan. 

One way to approach this question is to decompose the percent overlap figure from above into the distributions of overlaps for each separate amino acid.

```python
from hmmhelperfuncs import *

aas = get_aas()
aas_to_overlaps = {aa:[] for aa in aas}

for idx, row in lcr_and_idrs.iterrows():
    idrs = [tuple(idr.split('_')) for idr 
            in row['IDR30'].split(';')]
    idr_pos = set()
    # Compute IDR positions and pool them
    [idr_pos.update(range(int(idr[0]), int(idr[1])+1)) for idr in idrs]
    
    aas_used = []
    for lcr in re.findall(query, row['LCRs']):
        aas_used.append(lcr[0])
        lcr_pos = ast.literal_eval(lcr[1])
        per_ovlp = len(set(lcr_pos).intersection(idr_pos))/float(len(lcr_pos))*100
        aas_to_overlaps[lcr[0]].append(per_ovlp)

to_plot = []
to_plot_aas = []
print("Number of LCRs for the following amino acids are too low\
(< 100) and are ignored:")
for aa in aas:
    if len(aas_to_overlaps[aa]) >= 100:
        to_plot.append(np.array(aas_to_overlaps[aa]))
        to_plot_aas.append(aa)
    else:
        print(aa, end=" ")
        
with sns.axes_style('ticks'):
    f = plt.figure(figsize=(8,4))
    ax = f.add_subplot(111)
    sns.despine(ax=ax)
    sns.offset_spines(ax=ax)
    sns.axlabel("","LCR-IDR percent overlap")
    sns.violinplot(to_plot, names=to_plot_aas, ax=ax, color="Set2", lw=0);
    f.savefig(config['output']['plots']+'lcr_idr_overlap_aa.png', transparent=True)
```

## Caveats

- CAST only predicts regions of singly-enriched amino acids. As a result, the amino acids in regions that are doubly-enriched (e.g. FG) are considered separately, which might lower the significance of both amino acids below the threshold. **It would be interesting to re-run this analysis, but using SEG**
- A hard-cut off of 30 amino acids was used for the disordered regions, this may be too strict.
- Cleavage of a precursor protein wasn't taken into account. So IDRs/LCRs may span a region that ends up in two separate proteins. (This is difficult problem).
