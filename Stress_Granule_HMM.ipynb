## Motivation

RNA granules, including stress granules and P-bodies, are dynamic, membrane-independent, phase-dense cytoplasmic structures that contain translationally-silenced host mRNPs and play an important role in the mRNA lifecycle, stress response, and viral infection. It has been known that most components of such granules have much shorter residence times than the granules themselves, but a physical explanation has arisen only recently. A growing body of evidence suggests that transient, low-affinity interactions between the low complexity regions (LCRs) and intrinsically disordered regions (IDRs) of granule-associated proteins act to maintain the granules. The meta-stable nature of these granules make them difficult to assess *in vivo* and reproduce realistically *in vitro*. Any systematic analysis of the granule-associating proteome is hindered by the lack of a reliable way of discerning, or at least predicting, if a given protein is likely to associate or target RNA granules. Recent works have shown that low complexity (LC) sequences seem to play an important role in the targeting of a protein to granules. For example, removing the LC-rich N-terminal of the DYRK3 kinase inhibits its ability to target stress granules, while the N-terminal, when expressed independently, forms granule-like structures *in vivo*. LC regions greatly vary in their physical properties so it is very likely that certain LCRs have a higher degree of granule-targeting effect. This would be similar to the Q/N-rich nature of prion domains (PrDs) that cause the pathological and irreversible protein aggregration characteric of prion diseases. Unlike for PrDs, there has not been a thorough analysis of the nature of SG/RG-targeting sequences. Such an analysis would allow for:

- Narrowing the search space in the human proteome for potential RNA granule-targeting proteins
- Identification of viral proteins that target viroplasms (which are thought be RNA-granule-like)

```python
import pandas as pd, numpy as np, yahmm as hmm
from mischelperfuncs import *
from numpy.lib.stride_tricks import as_strided as ast
import json
config = json.load(open("config.json"))
```

## Building the Model

First, let's load the complete, reviewed human proteome downloaded from Uniprot and calculate background amino acid frequencies.

```python
human_proteome = pd.read_csv(config['rawdata']['human'], index_col=0)
human_proteome.drop(human_proteome.iloc[-1,:], inplace=True)
background_freqs = get_freqs(''.join(human_proteome['SEQ'].tolist()), pseudocount=0)
```

### Determining the AA compositional bias in the seed sequences

Next, the compositional bias of the seed sequences will need to be determined. The length of the sequences vary and each should be weighted equally so the amino acid frequencies will be length normalized before they are combined. A pseudocount of 1 was added to all amino acids to reduce effects of overfitting (explore other options here).

```python
seed_seqs = human_proteome[human_proteome.index.isin(['P31483', 'Q01085'])]

sgnr = {'P31483':(290,387),'Q01085':(281,375)} # Regions from Gilks et al 2004

# Stress granule targeting sequences
sgtr_seqs = []

for index, row in seed_seqs.iterrows():
    start, end = sgnr[index]
    sgtr_seqs.append(row['SEQ'].replace(' ', '')[start-1:end])
    print(get_freqs(sgtr_seqs[-1]))

normed_seed_seqs = get_normed_freqs(sgtr_seqs)

print("Amino acid composition change background --> stress granule targeting")
for aa in get_aas():
    print('%s : %.2f%%'%(aa, normed_seed_seqs[aa]/background_freqs[aa]*100))
    
cPrDs = {"Sup3p": "NQGNNQ QNYQQYSQNG NQQQGNNRYQ GYQAYNAQAQ PAGGYYQNYQ\
GYSGYQQGGY QQYNPDAGYQ QQYNPQGGYQ QYNPQGGYQQ QFNPQGGRGN\
YKNFNYNNNL QGYQAGFQPQ SQGMSLNDFQ KQQKQ", "Rnq1p":"QGQGQGQGQGQGQGQGQGQGSFTALASLASSFMNSNNNNQQGQNQSSGGSSFGALASMAS\
SFMHSNNNQNSNNSQQGYNQSYQNGNQNSQGYNNQQYQGGNGGYQQQQGQSGGAFSSLAS\
MAQSYLGGGQTQSNQQQYNQQGQNNQQQYQQQGQNYQHQQQGQQQQQGHSSSFSALASMA\
SSYLGNNSNSNSSYGGQQQANEYGRPQQNGQQQSNEYGRPQYGGNQNSNGQHESFNFSGN\
FSQQNNNGNQ", "Ure2p":"MNNNGNQVSNLSNALRQVNIGNRNSNTTTDQSNINFEFSTGVNNNNNNNSSSNNNNVQNN\
NSGRNGSQNNDNENNIKNTLEQHRQQQQ", "New1":"MPPKKFKDLNSFLDDQPKDPNLVASPFGGYFKNPAADAGSNNA\
SKKSSYQQQRNWKQGGNYQQGGYQSYNSNYNNYNNYNNYNNYNNYNNYNKYNGQGYQKSTYKQSAVTPNQSGTPTPSASTTSLTSLNEKLSNLELTPISQFLSKIPECQS"}

normed_cprd_seqs = get_normed_freqs(cPrDs.values())
print("Amino acid composition change stress granule targeting --> core prion related domain")
for aa in get_aas():
    print('%s : %.2f%%'%(aa, normed_cprd_seqs[aa]/normed_seed_seqs[aa]*100))
```

There is a large enrichment of glutamine, tryptophan (!), tyrosine, glycine, methionine, asparagine, and proline in the prion-like domains of TIA-1 and TAR. There is also a lack of charged residues and hydrophobic residues. 

```python
background = hmm.State(hmm.DiscreteDistribution(dict(background_freqs)), 
                       name='B')
sgt_like = hmm.State(hmm.DiscreteDistribution(dict(normed_seed_seqs)), 
                     name='N')
cprds = hmm.State(hmm.DiscreteDistribution(dict(normed_cprd_seqs)),
                  name = 'P')

model = hmm.Model('stress_granule_targeting_screen')

model.add_state(background)
model.add_state(sgt_like)
model.add_state(cprds)
```

### Initial Transition probabilities

It's much more likely that a given protein sequence will have a background amino acid composition; determining exact probabilities for these values is difficult because the prevalence of stress granule-targeting sequences within the proteome is unknown. I'll start with the values used by Alberti et al 2009, but this could be refined in the future (see Baum-Welch method).

```python
model.add_transition(model.start, sgt_like, 0.025)
model.add_transition(model.start, cprds, 0.025)
model.add_transition(model.start, background, 0.95)
```

### Heterogeneous state transition probabilities

```python
model.add_transition(background, sgt_like, 0.0001)
model.add_transition(background, cprds, 0.0001)
model.add_transition(sgt_like, background, 0.02)
model.add_transition(sgt_like, cprds, 0.02)
model.add_transition(cprds, background, 0.02)
model.add_transition(cprds, sgt_like, 0.02)
```

### Homogeneous state transition probabilities

```python
model.add_transition(background, background, 0.9998)
model.add_transition(sgt_like, sgt_like, 0.96)
model.add_transition(cprds, cprds, 0.96)
```

Finalizing the model

```python
model.bake(verbose=True)
```

## Running the Model

```python
sgt_like_amt = []

window = 50

aas = get_aas()

indices = {state.name : i for i,state in enumerate(model.states)}

for index, row in human_proteome.iterrows():
    seq = [char for char in row['SEQ'] if char in aas]

    viterbi_path = model.viterbi(seq)[1]

    viterbi_pred = np.array([state[1].name for state in viterbi_path[1:]]) == 'N'
    starts = np.nonzero(viterbi_pred & ~np.roll(viterbi_pred, 1))[0]
    ends = np.nonzero(viterbi_pred & ~np.roll(viterbi_pred, -1))[0]
    sgtr_regions = [(i, j) for i, j in zip(starts, ends) if j+1 >= i + 30]

    ems = np.exp(model.forward_backward(list(seq))[1])
    probs = ems/np.sum(ems, axis=1)[:, np.newaxis]
    out = probs[:, indices['N']]
    padding = np.zeros(window - 1)
    padded_out = np.concatenate([padding,out])
    strided = ast(padded_out,shape = (len(padded_out) + 1 - window, window),
                  strides = padded_out.strides * 2)
    scores = [0]
    for i,j in sgtr_regions:
        scores.append(strided.sum(1)[i:j+1].max())
    sgt_like_amt.append(max(scores))

human_proteome['SGTSIM'] = sgt_like_amt
sgt_prots = human_proteome[human_proteome['SGTSIM'] > 0].sort('SGTSIM', ascending=False)
sgt_file = config['rawdata']['human'].split('.csv')[0] + "_sg_targeting.csv"
config['processed']['human']['sgt'] = sgt_file
sgt_prots.to_csv(sgt_file, index=False)
json.dump(config, open('config.json', 'w'), indent=4)
```
